Menyesuaikan folder yang ada pada data train dan validation dengan video_id nya

```python
import os
import shutil

def func(csv_path, folder_path):

    data = pd.read_csv(csv_path)

    valid_video_ids = set(data['video_id'].astype(str))

    semua_folder = os.listdir(folder_path)

    for i in semua_folder:

        folder_path_full = os.path.join(folder_path,i)

        if i not in valid_video_ids:
            
            if os.path.isdir(folder_path_full):

                shutil.rmtree(folder_path_full)
                print(f'Folder {folder_path_full} dihapus')
            else:
                print(f'Folder {folder_path_full} bukan folder, dilewati')
        else:
            print(f"Folder {folder_path_full} sesuai, dipertahankan.")
```

```python
root = "D:/TUBES DEEP LEARNING ANJAY" #Sesuaikan dengan file root kalian
```

```python
func(os.path.join(root,"Train.csv"), os.path.join(root,"Train"))
```

```python
func(os.path.join(root,"Validation.csv"), os.path.join(root,"Validation"))
```

Mengambil dari data train untuk dijadikan data test

```python
import os
import shutil
import pandas as pd

# Path awal dan tujuan
data_train_path = "D:/TUBES DEEP LEARNING ANJAY/Train"
data_test_path = "D:/TUBES DEEP LEARNING ANJAY/data_test"
train_csv_path = "D:/TUBES DEEP LEARNING ANJAY/Train.csv"
test_csv_path = os.path.join(data_test_path, "test.csv")

# Pastikan folder data_test ada
if not os.path.exists(data_test_path):
    os.makedirs(data_test_path)

# Load train.csv
train_data = pd.read_csv(train_csv_path)

# Ambil 10 video_id (bisa acak atau terurut)
selected_video_ids = train_data['video_id'].unique()[:10]  # Pilih 10 video_id pertama

# Pindahkan folder video_id ke dalam data_test
for video_id in selected_video_ids:
    src_folder = os.path.join(data_train_path, str(video_id))
    dst_folder = os.path.join(data_test_path, str(video_id))
    if os.path.exists(src_folder):
        shutil.move(src_folder, dst_folder)  # Pindahkan folder

# Buat test.csv hanya untuk video_id yang terpilih
test_data = train_data[train_data['video_id'].isin(selected_video_ids)]

# Pastikan kolom `label` dan `label_id` ada
if 'label' not in test_data.columns or 'label_id' not in test_data.columns:
    test_data['label'] = None
    test_data['label_id'] = None

# Simpan test.csv
test_data[['video_id', 'label', 'label_id']].to_csv(test_csv_path, index=False)

# Perbarui train.csv dengan menghapus video_id yang sudah dipindahkan
train_data = train_data[~train_data['video_id'].isin(selected_video_ids)]
train_data.to_csv(train_csv_path, index=False)

print(f"Data test berhasil dibuat di folder {data_test_path} dengan file test.csv")
print(f"train.csv telah diperbarui dengan menghapus video_id yang dipindahkan.")
```
